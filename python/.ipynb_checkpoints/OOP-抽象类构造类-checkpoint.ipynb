# 抽象类
- 可以被继承
- 如果想要实例化必须继承所有的方法

```python
# 抽象类的实现
# 抽象类没有具体方法实现,相当于Java的接口
import abc

# 声明一个类并使用当前类的元类
class Human(metaclass = abc.ABCMeta):
    
    #定义一个抽象类方法
    @abc.abstractmethod
    def somking(self):
        pass
    
    #定义一个类抽象方法
    @abc.abstractclassmethod
    def drink(self):
        pass
    
    #定义一个静态的抽象方法
    @abc.abstractstaticmethod
    def play(self):
        pass
    
    # 也可包含正常的方法
    def sleep(self):
        print('正在睡觉...')
        
```

# 构造类
- 一个类和各种方法组合使用 组装类
- 可以定义类核函数,然后自己通过类直接赋值
- 可以通过MethodType实现  MethodType(fun , class)
- 

```python
# 自己组装函数001

class A():
    pass
def say(self):
    print('sleeping...')
    
say(9)
A.say = say
a = A()
a.say()
```

```python
#函数名可以当成变量赋值给别人 
# JavaScript亲兄弟
def sayHello(name):
    print("{0},要来一发嘛".format(name))
    
s = sayHello
s('蛋蛋')

help(type)
```

```python
# 自己组装函数002
from types import MethodType
class A():
    pass
def say(self):
    print('sleeping...')
    
say(9)
a = A()
#实例化后再要赋值方法是不被允许的
# a.say = say

# 通过MethodType来绑定到一块儿
a.say = MethodType(say,A)
a.say()

help(MethodType)
```
